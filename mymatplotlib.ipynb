```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from matplotlib.colors import Normalize
from matplotlib.markers import MarkerStyle
from matplotlib.text import TextPath
from matplotlib.transforms import Affine2D
```

```python
array = [0.1, 1, ]

plt.plot(array)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height
ax.plot(array)
ax.set_title('Title')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
```

```python
# Define the range for the cube
size = 5  # Size of the cube, you can adjust this
x = np.linspace(0, size, 10)
y = np.linspace(0, size, 10)
z = np.linspace(0, size, 10)

# Create meshgrid
X, Y, Z = np.meshgrid(x, y, z)

# Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of points in the cubic grid
ax.scatter(X, Y, Z, color='b')

# Set labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Cube Plot')

plt.show()
```

```python
x = np.linspace(0, 10, 100)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(x, np.sin(x), label='sin(x)', color='b')
axs[0, 0].set_title('Sine Wave')

axs[0, 1].plot(x, np.cos(x), label='cos(x)', color='r')
axs[0, 1].set_title('Cosine Wave')

axs[1, 0].plot(x, np.tan(x), label='tan(x)', color='g')
axs[1, 0].set_title('Tangent Wave')
axs[1, 0].set_ylim(-10, 10)

axs[1, 1].scatter(x, np.sin(x), label='scatter', color='purple')
axs[1, 1].set_title('Scatter Plot')

for ax in axs.flat:
    ax.legend()
    ax.grid(True)

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

```python
theta = np.linspace(0, 2 * np.pi, 100)
r = np.abs(np.sin(theta))  # Radius as a function of angle

plt.polar(theta, r)
plt.title('Polar Plot')
plt.show()
```

```python
SUCCESS_SYMBOLS = [
    TextPath((0, 0), ""),
    TextPath((0, 0), ""),
    TextPath((0, 0), ""),
]

N = 25
np.random.seed(42)
skills = np.random.uniform(5, 80, size=N) * 0.1 + 5
takeoff_angles = np.random.normal(0, 90, N)
thrusts = np.random.uniform(size=N)
successful = np.random.randint(0, 3, size=N)
positions = np.random.normal(size=(N, 2)) * 5
data = zip(skills, takeoff_angles, thrusts, successful, positions)

cmap = plt.colormaps["plasma"]
fig, ax = plt.subplots()
fig.suptitle("Throwing success", size=14)
for skill, takeoff, thrust, mood, pos in data:
    t = Affine2D().scale(skill).rotate_deg(takeoff)
    m = MarkerStyle(SUCCESS_SYMBOLS[mood], transform=t)
    ax.plot(pos[0], pos[1], marker=m, color=cmap(thrust))
fig.colorbar(plt.cm.ScalarMappable(norm=Normalize(0, 1), cmap=cmap),
             ax=ax, label="Normalized Thrust [a.u.]")
ax.set_xlabel("X position [m]")
ax.set_ylabel("Y position [m]")

plt.show()
```
